```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.random.randint(1000, size=50)
b = np.random.randint(1000, size=50)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
color_array= ["r" if (x+y)%2==0 else 'b' for x,y in zip(a,b)]
ax.scatter(a, b, c = color_array)
```

```python
x = np.arange(0, 6)
y1 = 3*x + 0.5
y2 = 5*np.sqrt(x)
```

```python
print y1
print y2
```

```python
plt.plot(x, y1, '-.')
plt.plot(x, y2, '-H')
plt.xscale('log')
plt.legend(['3x+0.5','5x^0.5'])
plt.show()
```

```python
barheights = [3,5,1]
barlabels = ['grapes', 'oranges', 'hockey pucks']
plt.bar(np.arange(len(barheights)), barheights)
plt.xticks(np.arange(len(barheights)), barlabels, rotation=0)
plt.show()
```
